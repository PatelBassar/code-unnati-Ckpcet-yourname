```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# x axis values
x = [1,2,3,7,9]
# corresponding y axis values
y = [2,4,1,9,5]
```

```python
# naming the x axis
plt.xlabel('sales')

# naming the y axis
plt.ylabel('profit')


# giving a title to my graph
plt.title('My first graph!')
# plotting the points
plt.plot(x, y )
```

```python

#subplots

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])


#plot 2:
z = np.array([1, 2, 3, 4])
w = np.array([5, 10, 30, 40])


plt.subplot(2, 2, 1)
plt.plot(x,y)

plt.subplot(2, 2, 4)
plt.plot(z,w)
plt.subplot(2, 2, 2)
plt.plot(x,z)
plt.subplot(2, 2, 3)
plt.plot(x,w)
```
